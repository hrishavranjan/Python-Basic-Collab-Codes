```python
import random
import matplotlib.pyplot as plt

# Objective function to maximize
def objective_function(x):
    return -x**2 + 10  # Max at x=0, value=10

# Hill Climbing Algorithm
def hill_climbing(start_x, max_iterations, step_size):
    current_x = start_x
    current_value = objective_function(current_x)

    x_history = [current_x]
    y_history = [current_value]

    for i in range(max_iterations):
        neighbor_x = current_x + random.uniform(-step_size, step_size)
        neighbor_value = objective_function(neighbor_x)

        if neighbor_value > current_value:
            current_x = neighbor_x
            current_value = neighbor_value

        x_history.append(current_x)
        y_history.append(current_value)

    return current_x, current_value, x_history, y_history

# Run the algorithm
start = random.uniform(-10, 10)
best_x, best_value, x_path, y_path = hill_climbing(start_x=start, max_iterations=100, step_size=0.5)

print(f"Starting point: {start}")
print(f"Best solution: x = {best_x:.4f}, value = {best_value:.4f}")

# Plotting the search path
x_vals = [x / 100.0 for x in range(-1000, 1000)]
y_vals = [objective_function(x) for x in x_vals]

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='Objective Function', color='blue')
plt.plot(x_path, y_path, marker='o', linestyle='--', color='red', label='Search Path')
plt.title('Hill Climbing - Search Path Visualization')
plt.xlabel('x')
plt.ylabel('Objective Function Value')
plt.legend()
plt.grid(True)
plt.show()
```
